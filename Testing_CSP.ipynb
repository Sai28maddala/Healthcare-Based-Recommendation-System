```python
'''
Now we need to labelEncode our y data because the machine learn binary language 
fast and more accurate so here we use labelEncoder from scikit-learn library
to transform the data

'''
```

```python
le = LabelEncoder()
le.fit(y)
Y = le.transform(y)
```

```python
# split the data 
X_test,X_train,y_test,y_train = train_test_split(X,Y,test_size =0.2 , random_state=42)
```

```python
X_test.shape,X_train.shape,y_test.shape,y_train.shape
```

```python
#Train Top Models
```

```python
import numpy as np
import pandas as pd
from scipy.sparse import issparse

# Convert to dense format if sparse
if issparse(X_train):
    X_train = X_train.toarray()
if issparse(X_test):
    X_test = X_test.toarray()

# Ensure that data is in float format
X_train = np.array(X_train, dtype=np.float64)
X_test = np.array(X_test, dtype=np.float64)

# Rerun your model training and testing code
models = {
      'SVC' : SVC(kernel='linear'),
      'Random Forest' : RandomForestClassifier(random_state=42, n_estimators=100),
      'KNeighbors' : KNeighborsClassifier(n_neighbors=5),
      'Gradient Boosting': GradientBoostingClassifier(random_state=42, n_estimators=100),
      'MultinomialNB' : MultinomialNB() 
}

for model_name, model in models.items():
    # Train model
    model.fit(X_train, y_train)
    # Test model
    predictions = model.predict(X_test)
    # Calculate accuracy
    accuracy = accuracy_score(y_test, predictions)
    # Calculate confusion matrix
    cm = confusion_matrix(y_test, predictions)
    # Print results
    print(f"{model_name} accuracy: {accuracy}")
    print(f"{model_name} confusion matrix:")
    print(np.array2string(cm, separator=', '))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# Sample true labels and predicted labels
y_true = np.array([0, 1, 0, 1, 0, 1, 1, 0])  # Replace with your true labels
y_pred = np.array([0, 0, 1, 1, 0, 1, 1, 0])  # Replace with your model's predictions

# Create a sample feature set (replace with your actual features)
X = np.random.rand(len(y_true), 5)  # Example feature set

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_true, test_size=0.2, random_state=42)

# Initialize and fit the Random Forest Classifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Generate the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Display the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.show()

# Calculate accuracy
accuracy = (cm[0][0] + cm[1][1]) / np.sum(cm)
print(f'Accuracy: {accuracy * 100:.2f}%')
```
